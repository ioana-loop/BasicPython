```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

## Question 1

Load the Global Superstore csv file into a dataframe and view the last 10 rows of the data

```python
#Solution:
df = pd.read_csv('Global Superstore.csv')
df.tail(10)
```

## Question 2

Report the mean of Sales and standard deviation of Quantity columns

```python
#Mean of Sales:
print('the mean of sales is', df['Sales'].mean())
```

```python
#S.D. of Quantity:
df['Quantity'].std()
```

## Question 3

Find the Category with highest average (mean) Sales

(Hint: Use the groupby function)

```python
#Code: 
q3 = df.groupby('Category').mean()['Sales'].reset_index()
q3
q3_sort = np.argsort(q3['Sales'])
q3_sort
q3_high = q3_sort[-1:]
```

```python
#Answer: 

print(np.unique(df['Category'][q3_high]))
```

## Question 4

Find the Sub-Category with maximum shipping cost

```python
#Code:
q4 = df.groupby('Sub-Category').max()['Shipping Cost']
q4
```

```python
q4_sort = np.argsort(q4)
q4_sort
```

```python
q4_max = q4_sort[-1:]
```

```python
#Answer:
print(np.unique(df['Sub-Category'][q4_max]))
```

## Question 5

Create a new column 'Cost' where Cost is Sales divided by Quantity

```python
#Code:

df['Cost'] = df['Sales']/df['Quantity']
df
```

## Question 6

Make a line plot and show 'Sub-Category' with respect to the Sum of their 'Profits'.

Your X axis should have Sub-Categories and Y axis should have the sum of profits.

Make sure you label the plot (Title, XLabel, YLabel)

Hint: Use .plot function

```python
#Code:

df.groupby('Sub-Category').sum()['Profit'].plot()
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.title('Sub-Category with respect to the Sum of their Profits')
degrees = 70

#look I found a cool new feature
plt.xticks(rotation=degrees)

plt.show()
```

## Question 7

Make a barplot of Order Priority with their Average Shipping Cost

Your X axis should have different Order Priorities and Y axis should have Shipping Cost

Make sure you label your axis

Hint: Use sns.barplot()

```python
#Code:

q7 = df.groupby('Order Priority').mean()['Shipping Cost'].reset_index()
print(q7)
```

```python
plt.figure(figsize = [10,10])
sns.barplot('Order Priority','Shipping Cost',data = q7)
plt.show()
```

## Question 8

Make a scatter plot showing Quantity (X Axis) and Profit (Y Axis)

Use alpha = 0.7 and color = blue

Hint: Use df.plot.scatter()

```python
#Code:
plt.figure(figsize = [10,10])
q8 = df.loc[:,['Quantity','Profit']]
plt.scatter(q8['Quantity'],q8['Profit'],color = 'blue', alpha=0.7)
plt.xlabel('Quantity')
plt.ylabel('Profits')
plt.title('Quantity v. Profit Data')
plt.grid()
plt.show()
```

## Question 9

Make a boxplot for showing Ship Mode and Shipping Cost

The X axis should be Ship Mode and Y axis should be the Shipping Cost

Hint: Use sns.boxplot()

```python
#Code:
sns.boxplot(x=df['Ship Mode'], y = df['Shipping Cost'], data = df)
plt.xlabel('Ship Mode')
plt.ylabel('Shipping Cost')
plt.title('Quantity v. Profit Data')
plt.show()

##add axes
```
